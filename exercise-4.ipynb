```r
library(xlsx)
library(ggplot2)
library(dplyr) # for renaming columns
library(reshape2)
library(tidyr)
df = read.xlsx2("Utentes.xlsx", sheetName = "Sheet1")
```

```r
df <- type.convert(df, as.is = TRUE)
```

```r
# scatter plot df with "Idade" in the x axis and "Colesterol" in the y axis
ggplot(df, aes(x = Idade, y = Colesterol)) +
  geom_point(color = "red") +
  geom_smooth(method = lm, linetype = "dashed") +
  theme_minimal()
```
